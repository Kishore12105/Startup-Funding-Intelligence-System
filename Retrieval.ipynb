```python
import json
from tqdm import tqdm
import chromadb
from chromadb.config import Settings
from sentence_transformers import SentenceTransformer
```

```python
with open("/content/funding_chunks_layered (1).json", "r") as f:
    chunks = json.load(f)

print("✅ Total chunks loaded:", len(chunks))
print("✅ Sample metadata:", chunks[0]["metadata"])
```

```python
embedding_model = SentenceTransformer(
    "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
)
```

```python
client = chromadb.Client(
    Settings(
        persist_directory="./chroma_store",
        anonymized_telemetry=False
    )
)
```

```python
collection = client.create_collection(
    name="startup_funding_knowledge"
)

print("✅ Collection created")
```

```python
documents = []
metadatas = []
ids = []

for i, ch in enumerate(chunks):
    documents.append(ch["text"])
    metadatas.append(ch["metadata"])
    ids.append(str(i))

collection.add(
    documents=documents,
    metadatas=metadatas,
    ids=ids
)

print("Vectors added:", collection.count())
```

```python
query = "How much funding can I get under Startup India Seed Fund Scheme?"

results = collection.query(
    query_texts=[query],
    n_results=5
)

for i, doc in enumerate(results["documents"][0]):
    print(f"\n--- Result {i+1} ---")
    print(doc[:300])
    print("Metadata:", results["metadatas"][0][i])
```

```python
results_core = collection.query(
    query_texts=[query],
    n_results=5,
    where={"layer": "core_funding"}
)

for i, doc in enumerate(results_core["documents"][0]):
    print(f"\n--- Core Result {i+1} ---")
    print(doc[:300])
    print("Metadata:", results_core["metadatas"][0][i])
```

```python
def retrieve_context(query, k=5):
    q = query.lower()

    if any(x in q for x in ["tamil", "kerala", "karnataka", "telangana"]):
        where = {"layer": "state_level"}
    elif any(x in q for x in ["how much", "amount", "eligibility"]):
        where = {"layer": "core_funding"}
    else:
        where = None

    results = collection.query(
        query_texts=[query],
        n_results=k,
        where=where
    )

    context = ""
    for doc, meta in zip(results["documents"][0], results["metadatas"][0]):
        context += f"\n[Source: {meta['source_file']}]\n{doc}\n"

    return context
```

```python
def generate_answer(query):
    context = retrieve_context(query)
    retrieved_chunks = retrieve_context(query)
    context = "\n\n".join(retrieved_chunks)

    lang = detect_language(query)

    if lang == "en":
        language_instruction = "Answer strictly in English."
    elif lang == "hi":
        language_instruction = "Answer strictly in Hindi."
    elif lang == "ta":
        language_instruction = "Answer strictly in Tamil."
    elif lang == "te":
        language_instruction = "Answer strictly in Telugu."
    elif lang == "ml":
        language_instruction = "Answer strictly in Malayalam."
    else:
        language_instruction = "Answer strictly in English."

    prompt = f"""
You are an AI assistant that helps startups understand funding schemes in India.

STRICT RULES:
- {language_instruction}
- Use ONLY the provided context.
- Do NOT hallucinate.
- Do NOT add extra information.
- If information is not available, say: "Information not found in official documents."

Context:
{context}

Question:
{query}

Answer:
"""


    model = genai.GenerativeModel("models/gemini-2.5-flash")
    return model.generate_content(prompt).text
```

```python
def multilingual_answer(query):
    lang = detect(query)

    # Translate to English if needed
    if lang != "en":
        query_en = genai.GenerativeModel("models/gemini-2.5-flash") \
            .generate_content(f"Translate to English:\n{query}").text
    else:
        query_en = query

    answer_en = generate_answer(query_en)

    # Translate back if needed
    if lang == "en":
        return answer_en
    else:
        final_answer = genai.GenerativeModel("models/gemini-2.5-flash") \
            .generate_content(f"Translate to {lang}:\n{answer_en}").text
        return final_answer
```
